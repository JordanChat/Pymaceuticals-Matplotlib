```python
# Dependencies and Setup
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"


# Read the mouse data and the study results
mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata_df,study_results_df, on='Mouse ID', how='left')
# Display the data table for preview
combined_df
```

```python
# Check the number of mice.
total_mice1 = combined_df["Mouse ID"].count()
total_mice1
```

```python
# Find any duplicate rows with the same Mouse ID's and Timepoints. 
combined_df2 = pd.DataFrame(combined_df)
combined_df2

#dropping duplicate rows
main_df = combined_df2.drop_duplicates(subset=['Mouse ID','Timepoint'])
main_df
```

```python
# Recheck the number of mice
total_mice2 = main_df["Mouse ID"].count()
total_mice2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, 
# standard deviation, and SEM of the tumor volume for each regimen.
summary_statistics = pd.DataFrame(main_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']}))
summary_statistics
```

```python
#putting every column into a count, i need Mouse ID count for number of experiments
summary_statistics2 = pd.DataFrame(main_df.groupby('Drug Regimen').count())
summary_statistics2.head()
```

```python
#seperating the Mouse ID amounts so I can use it to chart my data
experiment_list = pd.DataFrame(summary_statistics2.loc[:, ["Mouse ID"]])
experiment_list

# experiment_df = pd.merge(experiment_list, summary_statistics, on='Drug Regimen')
# experiment_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of datapoints for each drug regimen using pandas.
# There should be a single bar per regimen


experiment_list.plot(kind="bar", color="#CA6F33", title= "Datapoints For Each Drug Regimen")
plt.legend()
plt.xlabel('Drug Regimen')
plt.ylabel('Experiments')

plt.tight_layout()
plt.show()
```

```python
# Generate identical bar plot using pyplot instead of pandas.
x_axis = np.arange(len(experiment_list))
plt.bar(x_axis, experiment_list["Mouse ID"], color="g", alpha=0.5)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, list(experiment_list.index), rotation="vertical")
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(experiment_list["Mouse ID"])+10)
plt.title("Datapoints For Each Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Experiments")

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
#first finding the counts of male and female
explode = (0.1,0)
colors = ["cornflowerblue", "plum"]
mice_gender_dist = main_df["Sex"].value_counts()
plt.figure()
mice_gender_dist.plot(kind="pie", explode=explode, colors=colors, autopct= '%1.1f%%', shadow=True)
plt.axis("equal")
plt.title("Female vs Male Distribution")
plt.show()
```

```python
# Generate identical pie plot using pyplot
colors = ["indianred", "olive"]
labels = ["Male", "Female"]
explode = (0.1, 0)
plt.figure()
plt.pie(mice_gender_dist.values, labels=labels, explode=explode, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("Female vs Male Distribution")
plt.show()
```
